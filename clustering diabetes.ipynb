```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:/Users/diabetes.csv')
data
```

```python
data = data.drop(["Outcome"], axis=1)
data
```

```python
x_array=np.array(data)
print(x_array)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_array)
x_scaled
```

```python
KMeans = KMeans(n_clusters = 3, random_state=123)
```

```python
KMeans.fit(x_scaled)
```

```python
print(KMeans.cluster_centers_)
```

```python
data['cluster'] = KMeans.labels_
```

```python
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
reduce_data=PCA(n_components=2).fit_transform(x_scaled)
results = pd.DataFrame(reduce_data,columns=['PCA 1', 'PCA 2'])
results
```

```python
output =plt.scatter(results['PCA 1'], results['PCA 2'], s = 100, c = data.cluster, marker = 'o', alpha = 1)

plt.title("Hasil Clustering")
plt.colorbar(output)
plt.show()
```
